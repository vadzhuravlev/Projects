# Support Vector Machines Project 


## The Data

For this project we will be using the famous [Iris flower data set](http://en.wikipedia.org/wiki/Iris_flower_data_set). 

The data set consists of 50 samples from each of three species of Iris (Iris setosa, Iris virginica and Iris versicolor), so 150 total samples. Four features were measured from each sample: the length and the width of the sepals and petals, in centimeters.

The three classes in the Iris dataset:

    Iris-setosa (n=50)
    Iris-versicolor (n=50)
    Iris-virginica (n=50)

The four features of the Iris dataset:

    sepal length in cm
    sepal width in cm
    petal length in cm
    petal width in cm

## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Get the data

```python
df = pd.read_csv("iris.csv")

df.info()
```

```python
df.head()
```

## Exploratory Data Analysis

```python
sns.set_style('whitegrid')
sns.pairplot(df, hue = "species")
```

Based off this pairplot: setosa class is high separable from other species.

```python
sns.kdeplot(df[df['species'] == 'setosa']['sepal_width'], df[df['species'] == 'setosa']['sepal_length'], cmap = "coolwarm", shade = True, shade_lowest = False)
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(df.drop('species', axis = 1), df['species'], test_size = 0.3)
```

# Train a Model

```python
from sklearn.svm import SVC

model = SVC()
model.fit(X_train, Y_train)
```

## Model Evaluation

```python
from sklearn.metrics import classification_report, confusion_matrix

predict_data_svc = model.predict(X_test)
print(classification_report(Y_test, predict_data_svc))
```

```python
print(confusion_matrix(Y_test, predict_data_svc))
```

Based off classification_report and confusion_matrix: model fits and predicts very well. We will try to use GridSearch for test can we improve this algorithm.

## GridSearch

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'C' : [0.1, 1, 10, 100, 1000], 'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 'kernel' : ['rbf']}
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3)
grid.fit(X_train, Y_train)
```

```python
predict_data_grid = grid.predict(X_test)
print(classification_report(Y_test, predict_data_grid))
```

```python
print(confusion_matrix(Y_test, predict_data_grid))
```

Comparing the results for SVC without and with GridSearch: the changes are small, so it doesn't make sense to complicate this model.
