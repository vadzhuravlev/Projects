# K Means Clustering Project 

For this project we will attempt to use KMeans Clustering to cluster Universities into the two groups, Private and Public.

## The Data

We will use a data frame with 777 observations on the following 18 variables.
* Private: a factor with levels No and Yes indicating private or public university.
* Apps: number of applications received.
* Accept: number of applications accepted.
* Enroll: number of new students enrolled.
* Top10perc: pct. new students from top 10% of H.S. class.
* Top25perc: pct. new students from top 25% of H.S. class.
* F.Undergrad: number of fulltime undergraduates.
* P.Undergrad: number of parttime undergraduates.
* Outstate: out-of-state tuition.
* Room.Board: room and board costs.
* Books: estimated book costs.
* Personal: estimated personal spending.
* PhD: pct. of faculty with Ph.D.â€™s.
* Terminal: pct. of faculty with terminal degree.
* S.F.Ratio: student/faculty ratio.
* perc.alumni pct. alumni who donate.
* Expend: instructional expenditure per student.
* Grad.Rate: graduation rate.

## Import Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

In this dataset seems to be a private school with a graduation rate of higher than 100%.

```python
data[data['Grad.Rate']>100]
```

Set that school's graduation rate to 100 so it makes sense. 

```python
data['Grad.Rate'] = data['Grad.Rate'].apply(lambda x: 100 if x > 100 else x)
```

```python
data[data['Grad.Rate']>100]
```

```python
sns.set_style('whitegrid')
grid = sns.FacetGrid(data, hue = "Private", height = 5, aspect = 2)
grid.map(plt.hist, "Grad.Rate", bins = 20, alpha = 0.5)
```

## K Means Cluster Creation

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters = 2)
model.fit(data.drop('Private', axis = 1).dropna())
```

Cluster center vectors:

```python
model.cluster_centers_
```

## Evaluation

Check the results labels with actually results using confusion matrix and classification report.

```python
data.dropna(inplace = True)
data['Cluster'] = model.labels_
```

```python
data.head()
```

```python
from sklearn.metrics import classification_report, confusion_matrix

confusion_matrix(data['Private'].apply(lambda x: 1 if x == 'Yes' else 0), data['Cluster'])
```

```python
df1 = data['Private'].apply(lambda x: 1 if x == 'Yes' else 0)
classification_report(df1, data['Cluster'])
```

Results are pretty good, because we just use features to cluster data.
